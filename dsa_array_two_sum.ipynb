two sum
- given: array of integers and a target. return indices of the two numbers such that they add up to the target.
- constraints:
  - exactly one solution
  - array is atleast 2 in length and max upto 10^4
  - each value in array and target are both less than +/- 10^9 to avoid integer overflow/hash collision in most languages. python does not need this constraint.
- brute force:  - o(n.n)= o(n^2) complexity
iterate over entire array for each index to sum up with next index and match sum to target. once you find a match, exit.
- optimized: use hashmap of value and index


```python

#add your implementation and run the tests with run all
#expect failures until implementation is optimized to pass each edge case

def twoSum(nums, target):
  return [0,0]

def test_twoSum():

    # Edge Case 1: Minimum array size (exactly 2 elements)

    assert twoSum([1, 2], 3) == [0, 1]
    assert twoSum([5, 7], 12) == [0, 1]
    assert twoSum([-1, -2], -3) == [0, 1]


    # Edge Case 2: Negative numbers

    assert twoSum([-3, 4, 3, 90], 0) == [0, 2]  # -3 + 3 = 0
    assert twoSum([-1, -2, -3, -4, -5], -8) in [[2, 4], [4, 2]]  # -3 + -5
    assert twoSum([1, -1, 0], 0) == [0, 1]  # 1 + (-1) = 0


    # Edge Case 3: Target is zero

    assert twoSum([0, 0], 0) == [0, 1]  # 0 + 0 = 0
    assert twoSum([-5, 5, 1], 0) == [0, 1]  # -5 + 5 = 0
    assert twoSum([3, -3, 2], 0) == [0, 1]


    # Edge Case 4: Duplicate numbers (but different indices)

    assert twoSum([3, 3], 6) == [0, 1]  # Same value, different indices
    assert twoSum([2, 5, 5, 11], 10) == [1, 2]  # Two 5's
    assert twoSum([1, 1, 1, 1], 2) == [0, 1]  # Multiple duplicates


    # Edge Case 5: Answer at the extremes (first/last indices)

    assert twoSum([1, 9, 2, 3], 10) == [0, 1]  # First two elements
    assert twoSum([1, 2, 3, 7], 8) == [0, 3]  # First and last
    assert twoSum([5, 1, 2, 7], 9) == [2, 3]  # Last two elements


    # Edge Case 6: Large numbers (near constraint limit)

    assert twoSum([1000000000, -1000000000, 5], 5) == [1, 2]  # 10^9
    assert twoSum([999999999, 1], 1000000000) == [0, 1]


    # Edge Case 7: Numbers that could mislead (need same element twice)

    # target = 10, nums = [5, 2, 3]
    # Cannot use nums[0] twice! Must find two different indices
    assert twoSum([5, 2, 8], 10) == [0, 2]  # 5 + 5 ≠ valid, use 2 + 8
    assert twoSum([6, 3, 4], 7) not in [[0, 0]]  # Can't use same index twice


    # Edge Case 8: Answer requires looking back (not sequential)

    assert twoSum([3, 2, 4], 6) == [1, 2]  # Not [0, 0]
    assert twoSum([2, 7, 11, 15], 9) == [0, 1]
    assert twoSum([2, 5, 11, 7], 9) == [0, 3]  # Need to skip middle elements


    # Edge Case 9: Mix of positive, negative, and zero

    assert twoSum([0, 4, 3, 0], 0) == [0, 3]  # Two zeros
    assert twoSum([-1, 0, 1, 2], 1) == [1, 2]  # 0 + 1 = 1
    assert twoSum([-5, 0, 5, 10], 5) == [1, 2]  # 0 + 5 = 5


    # Edge Case 10: Large array

    large_array = list(range(10000))
    large_array.append(5000)  # 10001 elements
    result = twoSum(large_array, 10000)
    assert large_array[result[0]] + large_array[result[1]] == 10000


    # Edge Case 11: Answer at beginning vs end

    assert twoSum([1, 2, 3, 4, 5], 9) == [3, 4]  # Last two
    assert twoSum([5, 4, 3, 2, 1], 9) == [0, 1]  # First two


    # Edge Case 12: Complement equals the number itself

    # target = 8, looking for 4 + 4 (must use two different 4's)
    assert twoSum([1, 4, 4, 2], 8) == [1, 2]
    assert twoSum([3, 3], 6) == [0, 1]

    print("All Two Sum edge case tests passed!")

```
